```python
with open('input.txt', 'r') as file:
    data = file.readlines()
    
data

seeds = data[0].split(':')[1].split()

data[1]
```

```python
headings = []

numbers = {}


current_heading = ""

for idx in data:
    #print(idx)
    this = idx.split()
    if this == [] or this == "" or this ==  "\n":
        continue
    for item in this:
        print(item)
        if item == "" or item == "\n":
            continue
        if item.isdigit():
            heading = current_heading
            if heading not in headings:
                headings.append(heading)
            current_heading = ""
            
            #print('is digit')
            if heading in numbers:
                numbers[heading] += item
            elif heading not in numbers and heading != "":
                numbers[heading] = [item]
        else:
            #print('is not digit')
            if item == "":
                continue
            current_heading += item.replace(':', '').strip() 
            #print(f"{current_heading} is current heading")
        #print('')
      
    
    
```
